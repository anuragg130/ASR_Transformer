```python
audioData = [] 
srate = [] 

parent_folder = r"C:\Users\Anurag\Desktop\ASR Transformer\UserLibri\audio_data\test-clean" # path to the parent folder

for root, dirs, files in os.walk(parent_folder):
    for filename in files:
        if filename.endswith(".flac"): # check if the file is a wav file
            file_path = os.path.join(root, filename) # create the full file path
            data, sampleRate = sf.read(file_path)
            audioData.append(data) 
            srate.append(sampleRate) 
            #plt.figure(figsize=(10,3)) 
            #lb.display.waveshow(data)
            #plt.title(os.path.join(root, filename)) # add the full path as title to the plot

#plt.show() # show all the plots
```

```python
for a, s in zip(audioData, srate):
    mfcc = librosa.feature.mfcc(y=a, sr=s) 
    print(mfcc)
```

```python
for a, s in zip(audioData, srate): 
    mfcc = librosa.feature.mfcc(y=a, sr=s) 
    mfccScaled = np.mean(mfcc.T, axis=0) 
    print(mfccScaled) 
```
